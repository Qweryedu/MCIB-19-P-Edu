```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.fftpack import fft
from scipy.signal import hann, butter, cheby1, lfilter
```

```python
def filtra(b,signal, a=1):
    filt1 = lfilter(b,a,signal)
    filt1 = lfilter(b,a,filt1[::-1])
    return filt1[::-1]
```

```python
def dB(datos):
    Mag = np.abs(datos)
    Mag = 20*np.log10(Mag)
    
    return Mag
```

```python
def abre(path, file_name):
    data = np.loadtxt(path+file_name, delimiter='\t', dtype='str')
    data = data[:-1,:-1].T.astype('float') #Elimina la ltima fila y columna que no contienen nada
    
    return data

path = '/home/omarpr/git/MCIB-19-P/eval/ME/data/P1/' #Cambiar dependiendo de la compu
file = 'Apnea1.txt'


signals = abre(path,file)

EKG = signals[1]
EMG = signals[3]

plt.figure('EKG',figsize=(35,15))
plt.plot(EKG, 'm')
plt.grid()
plt.title('EKG', size=(55)).set_color(color='hotpink')
plt.figure('EMG', figsize=(35,15))
plt.title('EMG', size=(55)).set_color(color='teal')
plt.plot(EMG, 'b')
plt.grid()
plt.show()
```

```python
def compfft(signal,orden):
    s_f = fft(signal)
    
    #Fc
    Fc1 = np.array([.7, 10])*2/200
    
    #Ventanas
    Wh = hann(signal.shape[0])
    bc,ac = cheby1(orden,0.1,Fc1, btype='bandpass')
    bb, ab = butter(orden,Fc1, btype='bandpass')
    print ( 'a: {}\nb: {}'.format(bc,ac) )
    #Filtramos
    s_fh = filtra(Wh,signal)
    s_fc = filtra(bc,signal,ac)
    s_fb = filtra(bb,signal,ab)
    
    #fft
    fh = fft(s_fh)
    fc = fft(s_fc)
    fb = fft(s_fb)
    
    #Magnitudes
    Ms = dB(s_f)
    Mh = dB(fh)
    Mc = dB(fc)
    Mb = dB(fb)
    
    #Graficas fft's
    plt.figure('FFT',figsize=(45,30))
    plt.title('FFT', fontsize=55).set_color(color='red')
    plt.grid()
    plt.plot(Ms[:len(Ms)//2], 'm')
    plt.plot(Mh[:len(Mh)//2], color='sienna')
    plt.plot(Mc[:len(Mc)//2], color='rebeccapurple')
    plt.plot(Mb[:len(Mb)//2], color='crimson')
    plt.xlabel('Frecuencia en muestras', fontsize=50).set_color(color='white')
    
    #Graficas seal
    plt.figure('Signal', figsize=(45,30))
    plt.title('Seales filtradas', fontsize=55).set_color(color='red')
    plt.grid()
    plt.plot(signal[:1000], 'm')
    #plt.plot(s_fh, color='sienna')
    plt.plot(s_fc[:1000], color='rebeccapurple')
    plt.plot(s_fb[:1000], color='crimson')
    plt.xlabel('Frecuencia en muestras', fontsize=50).set_color(color='white')
    
```

```python
compfft(EKG,2)
```
