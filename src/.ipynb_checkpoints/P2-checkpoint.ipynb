```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.fftpack import fft
from scipy.signal import hann, butter, cheby1, lfilter
```

```python
def filtra(b,signal, a=1):
    filt1 = lfilter(b,a,signal)
    filt1 = lfilter(b,a,filt1[::-1])
    return filt1
```

```python
def abre(path, file_name):
    data = np.loadtxt(path+file_name, delimiter='\t', dtype='str')
    data = data[:-1,:-1].T.astype('float') #Elimina la ltima fila y columna que no contienen nada
    
    return data

path = '/home/edu/Documentos/GitHub/MCIB-19-P-Edu/data/P1/' #Cambiar dependiendo de la compu
file = 'Apnea1.txt'


signals = abre(path,file)

EKG = signals[1]
EMG = signals[3]

plt.figure('EKG',figsize=(35,15))
plt.plot(EKG, 'm')
plt.grid()
plt.title('EKG', size=(55)).set_color(color='hotpink')
plt.figure('EMG', figsize=(35,15))
plt.title('EMG', size=(55)).set_color(color='teal')
plt.plot(EMG, 'b')
plt.grid()
plt.show()
```

```python
def compfft(signal):
    s_f = fft(signal)
    
    #Ventanas
    Wh = hann(signal.shape[0])
    bc,ac = cheby1(2,0.1,5,fs=200)
    bb, ab = butter(2,5, fs=200)
    print ( 'a: {}\b: {}'.format(bc,ac) )
    #Filtramos
    s_fh = filtra(Wh,signal)
    s_fc = filtra(bc,signal,ac)
    s_fb = filtra(bb,signal,ab)
    
    #fft
    fh = fft(s_fh)
    fc = fft(s_fc)
    fb = fft(s_fb)
    
    #Graficas
    plt.figure(figsize=(35,20))
    plt.grid()
    plt.plot(s_f, 'm')
    plt.plot(fh, color='sienna')
    plt.plot(fc, color='rebeccapurple')
    plt.plot(fb, color='crimson')
```

```python
compfft(EMG)
```
